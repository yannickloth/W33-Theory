## Cross Ratios & Möbius Transformations

The cross ratio of four points \(A,B,C,D\) on the projective line is

\[
(A,B;C,D) = \frac{(A-C)(B-D)}{(A-D)(B-C)}
\]

It is invariant under fractional linear (Möbius) transformations of the
form
\(z\mapsto\frac{az+b}{cz+d}\).  Over finite fields, the same formula
works with field arithmetic and PGL(2,q) replaces PSL(2,C).

We can use cross ratios to characterize configurations of four points up
to projective equivalence, and these invariants might encode ratios of
eigenvalues or other mass ratios.


````python


```python
# quick demonstration of cross ratio invariance using the new utility module
from scripts.w33_cross_ratio import cross_ratio, random_mobius, apply_mobius_to_list

# real numbers
pts = [0, 1, 2, 3]
print('cross ratio of', pts, cross_ratio(*pts))
a, b, c, d = random_mobius()
trans = apply_mobius_to_list(pts, a, b, c, d)
print('after mobius', (a,b,c,d), '->', trans)
print('cross ratio after', cross_ratio(*trans))

# finite field example
print('GF(7) cross ratio', cross_ratio(1, 2, 3, 4, field=7))
a, b, c, d = random_mobius(field=7)
trans = apply_mobius_to_list([1,2,3,4], a, b, c, d)
print('GF7 mobius', (a,b,c,d), '->', trans)
print('cross ratio after GF7', cross_ratio(*trans, field=7))
```
````
