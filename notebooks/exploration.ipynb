## Avenue 2: Numerical Simulation

Run a quick scan of the combined CKM/mass landscape and visualize the Pareto
frontier.

```python
# sample a handful of random points and plot ck_err vs mass_err
from combined_ckm_mass_landscape import build_yukawa_tensor, analyze_active_subspace, random_active_points

T = build_yukawa_tensor()
rank, Vh = analyze_active_subspace(T)
pts = random_active_points(T, Vh, rank, 2000, weight_mass=1.0)
ck = [pt['ck_err'] for pt in pts]
mass = [pt['mass_err'] for pt in pts]
import matplotlib.pyplot as plt
plt.scatter(ck, mass, s=5)
plt.xlabel('CKM error'); plt.ylabel('mass error'); plt.title('Random samples');
plt.show()
```


## Avenue 3: Symbolic Computation

Attempt to identify closed‑form expression for the CKM error floor and derive
inequalities constraining the Pareto frontier.


```python
# use sympy to approximate the constant 0.3066628615989846
import sympy as sp
v=sp.nsimplify(0.3066628615989846, [sp.sqrt(2), sp.sqrt(3), sp.pi])
print('nsimplify gives', v)

# maybe try higher precision rational approximation
sp.nsimplify(sp.N(0.3066628615989846,50), [sp.sqrt(2), sp.sqrt(3), sp.pi, sp.log(2)])
```
